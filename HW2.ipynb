Exercise 2:

```python
x = Piecewise((0,t<-1), (t+1, t<=0), (1, t<=1), (0, true))
plot(x, (t, -5, 5))
```

```python
xnt = x.subs(t, -t)

#x(t) even
xe = 0.5 * (x + xnt)

plot (xe, (t, -5, 5))
```

```python
#x(t) odd
xo = 0.5 * (x - xnt)

plot (xo, (t, -5, 5))
```
