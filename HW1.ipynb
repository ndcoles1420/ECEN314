```python
#Import libraries
from math import *
from sympy import *

# i) make a piecewise function
t = symbols('t')
x = Piecewise((0, t<-1), (0, t>=1), (1, t>=0), (t+1, True))

# plot x(t) on the range [-2,2]
plot(x, (t, -2, 2))
```

```python
#y1(t) = x(-t)
y1 = x.subs(t,-t)
plot(y1, (t, -2, 2))
```

This plot has the same general shape as the x(t) plot, but it is flipped with the f(t) axis at the center. It is like a mirror image.

```python
#y2(t) = x(t-1)
y2 = x.subs(t, t-1)
plot(y2, (t, -2, 2))
```

This plot has the same shape as the x(t) plot, but it has shifted over to the right by one.

```python
#y3(t) = x(t+1)
y3 = x.subs(t, t+1)
plot(y3, (t, -2, 2))
```

This plot has the same shape as the x(t) plot, but it has shifted over to the left by one.

```python
#y4(t) = x(2t)
y4 = x.subs(t, 2*t)
plot(y4, (t, -2, 2))
```

This plot is the x(t) squeezed together. It shinks by a factor of 2, or is half as wide as the x(t) plot.

```python
#y5(t) = x(t/2)
y5 = x.subs(t, t/2)
plot(y5, (t, -2, 2))
```

The plot for this graph is the x(t) plot stretched out. It specifcally expands by a factor of 2.
